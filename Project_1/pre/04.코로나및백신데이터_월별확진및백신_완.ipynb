# 코로나 백신 현황

```python
import pandas as pd
import numpy as np
```

```python
co_vc = pd.read_csv('./Data/corona/서울특별시 코로나19 백신 예방접종 현황.csv', encoding='cp949')
co_vc
```

```python
co_vc.columns
```

```python
list = ['접종일', '당일 1차접종자 수', '1차접종률(%)', '당일 2차접종자 수',
       '2차접종률(%)', '당일 추가접종자 수', '추가접종률(%)']
co_vc_01 = co_vc[list]
co_vc_01
```

```python
co_vc_02['당일 추가접종자 수'] = co_vc_02['당일 추가접종자 수'].astype('int')
co_vc_02
```

# 접종자수별

```python
vc_count = co_vc_02[['접종일', '당일 1차접종자 수', '당일 2차접종자 수', '당일 추가접종자 수']]
vc_count
```

```python
vc_count.info()
```

```python
vc_count.columns = ['일자','1차접종자수', '2차접종자수', '추가접종자수']
vc_count
```

```python
vc_count.to_excel('./Data/가공완료/[백신]서울백신접종자수(일별).xlsx')
```

```python
from tqdm import tqdm
a = []
for i in tqdm(range(0, 301)):
    a.append(vc_count.loc[i, '일자'][:7])
a_1 = np.array(a).reshape(-1,1)
vc_count['월'] = a_1
vc_count
```

```python
vc_df = vc_count[['월', '1차접종자수', '2차접종자수', '추가접종자수']]
vc_df
```

```python
vc_df_gr = vc_df.groupby('월')[['1차접종자수', '2차접종자수', '추가접종자수']].median()
vc_df_gr1 = pd.DataFrame(vc_df_gr).reset_index()
vc_df_gr2 = vc_df_gr1.replace(0, np.NaN)
vc_df_gr3 = vc_df_gr2[:9]
vc_df_gr3
```

```python
vc_df_gr3.to_excel('./Data/가공완료/[백신]서울백신접종자수(월별).xlsx')
```

```python
import matplotlib.pyplot as plt
from matplotlib import rc, font_manager
rc('font', family = 'malgun gothic')
```

```python
plt.figure(figsize = (12,8))
vc_df_gr3.plot()
plt.show
```
