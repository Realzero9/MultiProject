```python
def creat_data(yy, mm):
    file_path = './Data/서울특별시 공공자전거 이용정보(시간대별)_20{0}년/서울특별시 공공자전거 이용정보(시간대별)_{0}.{1}.csv'.format(yy,mm)
    try:
        f = pd.read_csv(file_path, encoding='utf-8')
    except:
        f = pd.read_csv(file_path, encoding='cp949')
    df = f[['대여일자', '대여소번호', '이용건수', '사용시간', '이동거리']]
    df.columns = ['대여일자', '대여소번호', '이용건수', '이동시간', '이동거리']
    return(df) 
```
