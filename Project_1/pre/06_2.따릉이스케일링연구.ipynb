```python
from tqdm import tqdm
```

```python
list = ['15', '16', '17']

DF = pd.DataFrame()

for i in tqdm(list):
    my = pd.read_excel('./Data/대여소별 반납내역(20{}년).xlsx'.format(i))
    my['반납 일자 / 월'] = my['반납 일자 / 월'].astype('str')
    my['월'] = my['반납 일자 / 월']
    DF = DF.append(my, ignore_index=True)
    
```

```python
DF
```

```python
for i in tqdm(range(0, 357749)):
    DF['월'][i] = DF['반납 일자 / 월'][i][:7]
```

```python
DF.columns = ['대여일자', '대여건수', '월']
DF
```

# 월 전처리

```python
DF_gr = DF.groupby('월')['대여건수'].sum()
DF_gr1 = pd.DataFrame(DF_gr)
DF_gr1 = DF_gr1.reset_index('월')
DF_gr1
```

```python
DF_gr1.to_excel('./Data/가공완료/월별대여건수(2015~17).xlsx')
```

```python
last = pd.read_excel('./Data/가공완료/월별대여건수(2018~21).xlsx', index_col='Unnamed: 0')
last
```

```python
by_DF = DF_gr1.append(last)
by_DF
```

```python
by_DF.to_excel('./Data/가공완료/[따릉이]월별대여건수(2015~).xlsx')
```
