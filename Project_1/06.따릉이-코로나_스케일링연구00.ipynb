```python
import pandas as pd
import numpy as np
```

```python
co = pd.read_csv('./Data/corona/서울특별시 코로나19 확진자 발생동향.csv', encoding='cp949')
co
```

```python
co_df = co[['서울시 기준일', '서울시 확진자', '서울시 추가 확진', '서울시 사망']]
co_df.info()
```

```python
for i in range(0, 743):
    co_df['서울시 기준일'][i] = co_df['서울시 기준일'][i][:-3].replace('.','-')
```

```python
for i in range(720, 743):
    co_df['서울시 기준일'][i] = co_df['서울시 기준일'][i].replace('20-','2020-')
```

```python
co_df['월'] = co_df['서울시 기준일']
for i in range(0, 743):
    co_df['월'][i] = co_df['서울시 기준일'][i][:7]
```

# 서울시추가확진자수만 표시

```python
co_gr = rona.groupby('월')['서울시 추가 확진'].sum()
co_gr
```

```python
co_gr_df = pd.DataFrame(co_gr).reset_index()
co_gr_df.columns = ['월', '서울시추가확진']
co_gr_df
```

```python
import matplotlib.pyplot as plt
from matplotlib import rc, font_manager
rc('font', family = 'malgun gothic')
```

```python
plt.figure(figsize = (12,8))
rona_gr.plot()
plt.legend()
plt.show
```

# 따릉이 가져오기

```python
by = pd.read_excel('./Data/서울따릉이월별대여건수(2018~21).xlsx')
by
```

```python
by = by.drop('Unnamed: 0', axis=1)
by
```
