```python
import pandas as pd
import numpy as np
from tqdm import tqdm
```

# 2. 함수 정의 : 각 csv파일 -> 데이터프레임화
 - 컬럼명 변경해서 return

```python
def create_data(yy, mm):
    file_path = './Data/서울특별시 공공자전거 이용정보(시간대별)_{0}.{1}.csv'.format(yy,mm)
    try:
        f = pd.read_csv(file_path, encoding='utf-8', engine='python')
        try:
            if '사용시간' in f:
                f['이동시간'] = f['사용시간']
                f = f.drop('사용시간', axis=1)
        except:
            pass
    except:
        f = pd.read_csv(file_path, encoding='cp949', engine='python')
        try:
            if '사용시간' in f:
                f['이동시간'] = f['사용시간']
                f = f.drop('사용시간', axis=1)
        except:
            pass
    df = f[['대여일자', '대여시간', '대여소번호', '대여소명', '대여구분코드', '성별', '연령대코드', '이용건수', '운동량',
       '탄소량', '이동거리', '이동시간']]
    return df
```

- 파일 모으기 : DF

```python
DF_20 = pd.DataFrame()
year = ['20']
for y in year:
    for m in tqdm(range(1, 13)):
        try:
            mm = str(m).zfill(2)
            temp = create_data(y, mm)
            DF_20 = pd.concat([DF_20, temp], ignore_index=True)
        except:
            print('에러'+y+mm)
            pass
```

```python
DF_20 = DF_20[['대여일자', '대여구분코드', '성별', '이용건수', '이동거리', '이동시간']]
DF_20
DF_20.to_csv('./Data/이용정보2020(cols).csv', encoding='utf-8', index=False)
```
